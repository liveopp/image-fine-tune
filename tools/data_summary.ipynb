# Stats about data

```python
from glob import glob
import json
from os import path
import pandas as pd
from PIL import Image
%pylab inline
```

### 定义数据集的目录 （根据情况修改）

```python
train_dir = "../tfclass/food_data/train"
val_dir = "../tfclass/food_data/val"
```

### 申明函数

```python


def print_dataset_stats(root_dir):

    train_img_fns = glob(path.join(root_dir, '*/*.jpg'))

    print('{} training images.'.format(len(train_img_fns)))
    df_train = pd.DataFrame([dict(label=s.split('/')[-2], fn=s) for s in train_img_fns])

    print('{} classes'.format(len(df_train.label.unique())))
    print('{} training images'.format(df_train.shape[0]))
    print('    dish images: {}'.format(df_train[df_train.label != '000000'].shape[0]))
    
def visualize_one_class(img_folder, limit=60):
    img_fns = glob(path.join(img_folder, '*.jpg'))[:limit]
    print('{} images to visualize'.format(len(img_fns)))
    figure(figsize=(12, 12))
    ncol = 6
    nrow = len(img_fns) / ncol + 1
    for i, img_fn in enumerate(img_fns):
        subplot(nrow, ncol, i+1)
        imshow(Image.open(img_fn))

def visualize_images(img_fns, limit=60):
    img_fns = img_fns[:limit]
    print('{} images to visualize'.format(len(img_fns)))
    figure(figsize=(12, 12))
    ncol = 6
    nrow = len(img_fns) / ncol + 1
    for i, img_fn in enumerate(img_fns):
        subplot(nrow, ncol, i+1)
        imshow(Image.open(img_fn))
```

### 打印数据集的基本信息

```python
print('----------------------------------------------')
print_dataset_stats(train_dir)
print('----------------------------------------------')
print_dataset_stats(val_dir)
print('----------------------------------------------')
```

### 可视化某一个种菜品的图片

```python
visualize_one_class(path.join(train_dir, '000001'))
```
